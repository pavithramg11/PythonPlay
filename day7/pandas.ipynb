```python
df["Sex"].unique()
```

```python
df["Age"] = df["Age"].apply(lambda x: x+5)
```

```python
def categorize(age):
    if age >= 18:
        return "Adult"
    else:
        return "Child"

df["AgeGroup"] = df["Age"].apply(categorize)
```

```python
df["AgeGroup"].value_counts()
```

```python
df[(df["Age"] > 20) & (df["Sex"] == "female")]["Name"]
```
