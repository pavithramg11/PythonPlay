```python
import csv

total = 0
count = 0

with open("data.csv", "r") as file:
    reader = csv.reader(file)
    
    next(reader)  # skip header
    
    for row in reader:
        score = int(row[2])   # convert to int
        total += score
        count += 1

average = total / count
print("Average Score:", average)
```
