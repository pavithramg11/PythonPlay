```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x, y)
plt.title("Line plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
categories = ["A", "B", "C"]
values = [10,20,15]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
data = [1,2,2,3,3,3,4]
plt.hist(data)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,5,8]
plt.scatter(x, y)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df["Age"].hist()
plt.show()
```

```python
plt.scatter(df["Age"], df["Pclass"])
plt.xlabel("Age")
plt.ylabel("Pclass")
plt.show()
```
